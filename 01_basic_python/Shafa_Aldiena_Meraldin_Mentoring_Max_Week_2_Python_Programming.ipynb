# Task 1. Calculate Average Sessions

**Task Descriptions**

- PacBook is a Social Media platform and they want to know what is their user Average Sessions and Session Duration for each data
- Given a three input for `load_time` (in minutes) and `exit_time` (in minutes)

- To calculate Session Duration, you can use this formula

$$
\text{Session}_\text{duration} = \text{exit}_\text{time} - \text{load}_\text{time}
$$

- To calculate Average Sessions, you can use this formula

$$
\text{Average}_\text{sessions} = \frac{\sum_{n=1}^{3}{\text{Session_duration}_{n}}}{\text{N_SESSIONS}}
$$

- But there are some constraints in this task
  - If the input for `load_time` and `exit_time` is less than zero or negative values, it will print the result **`"Not a Valid Input"`**

**Input - 1**

```py
load_time_1 = 5
exit_time_1 = 10

load_time_2 = 2
exit_time_2 = 15.5

load_time_3 = 1
exit_time_3 = 22.3
```

**Output - 1**

```py
Session 1 Duration (in minutes): 5
Session 2 Duration (in minutes): 13.5
Session 3 Duration (in minutes): 21.3
Average Session Duration (in minutes): 13.27
```

**Input - 2**

```py
load_time_1 = 0
exit_time_1 = 0

load_time_2 = 0
exit_time_2 = 0

load_time_3 = 0
exit_time_3 = 0
```

**Output - 2**

```py
Session 1 Duration (in minutes): 0
Session 2 Duration (in minutes): 0
Session 3 Duration (in minutes): 0
Average Session Duration (in minutes): 0
```

**Input - 3**

```py
load_time_1 = 3
exit_time_1 = 7

load_time_2 = -2
exit_time_2 = 5

load_time_3 = 3
exit_time_3 = 156
```

**Output - 3**

```py
Not a Valid Input
```

````python
# Write your code here

#input
load_time_1 = 5
exit_time_1 = 10

load_time_2 = 0
exit_time_2 = 0

load_time_3 = 3
exit_time_3 = 156

# jawaban

#session duration (exit time - load time)
duration_1 = exit_time_1 - load_time_1
duration_2 = exit_time_2 - load_time_2
duration_3 = exit_time_3 - load_time_3

#amount of session
n = 3

#average
average_duration = (duration_1 + duration_2 + duration_3)/n

#show result

#duration 1
if duration_1 < 0 or exit_time_1 < 0 or load_time_1 < 0:
    print("Not a Valid Input")
else :
  print(f"Session 1 Duration (in minutes): {duration_1}")

#duration 2
if duration_2 < 0 or exit_time_2 < 0 or load_time_2 < 0:
  print("Not a Valid input")
else :
  print(f"Session 2 Duration (in minutes): {duration_2}")

#duration 3
if duration_3 < 0 or exit_time_3 < 0 or load_time_3 < 0:
  print("Not a Valid input")
else :
  print(f"Session 3 Duration (in minutes): {duration_3}")

#average
if duration_1 < 0 or duration_2 < 0 or duration_3 < 0 or exit_time_1 < 0 or load_time_1 < 0 or exit_time_2 < 0 or load_time_2 < 0 or exit_time_3 < 0 or load_time_3 < 0 :
  print("Average Session Duration (in minutes): Not a Valid Input")
else :
 print(f"Average Session Duration (in minutes): {average_duration}")
````

````python
# Write your code here

#Insert Input
from_unit = "celcius"
to_unit = "fahrenheit"
temp = 20

#1. Celcius to Kelvin
ck = temp + 273.15

#2. Kelvin to Celcius
kc = temp - 273.15

#3. Fahrenheit to Kelvin
fk = 5/9 * (temp - 32) + 273.15

#4. Kelvin to Fahrenheit
kf = 9/5 * (temp - 273.15) + 32

#5. Celcius to Fahrenheit
cf = (9/5) * temp + 32

#6. Fahrenheit to Celcius
fc = 5/9 * (temp - 32)

if from_unit == "celcius" and to_unit == "kelvin":
  print(f"{temp} {from_unit} is equal to {ck} {to_unit}")

elif from_unit == "kelvin" and to_unit == "celcius":
  print(f"{temp} {from_unit} is equal to {kc} {to_unit}")

elif from_unit == "fahrenheit" and to_unit == "kelvin":
  print(f"{temp} {from_unit} is equal to {fk} {to_unit}")

elif from_unit == "kelvin" and to_unit == "fahrenheit":
  print(f"{temp} {from_unit} is equal to {kf} {to_unit}")

elif from_unit == "celcius" and to_unit == "fahrenheit":
  print(f"{temp} {from_unit} is equal to {cf} {to_unit}")

elif from_unit == "fahrenheit" and to_unit == "celcius":
  print(f"{temp} {from_unit} is equal to {fc} {to_unit}")

else :
  print("Not a Valid Input")
````

# Task 4. Calculate Pajak Penghasilan (Simplified)

Category: Hard

---

**Task Descriptions**
- According to Pemotongan Pajak Penghasilan PPh Pasal 21 Tahun 2016, we can calculate our income tax as (we'll write in Bahasa Indonesia for avoiding missinterpretation)

- **Penghasilan tidak kena pajak (PTKP)**
  - Penghasilan Rp 54.000.000 setahun untuk wajib pajak.
  - Tambahan Rp 4.500.000 diberikan jika wajib pajak menikah.
  - Tambahan Rp 4.500.000 lagi jika punya 1 orang anak atau Rp 9.000.000 jika punya 2 orang anak. Apabila jumlah tanggungan lebih dari itu, maka tetap dihitung 3 orang tanggungan.

- **Tarif Pajak Penghasilan**
  - 5% untuk penghasilan kena pajak (PKP) sampai dengan 60 Juta
  - 15% untuk PKP di atas 60 juta sampai dengan 250 Juta
  - 25% untuk PKP di atas 250 juta sampai dengan 500 juta
  - 30% untuk PKP di atas 500 juta sampai dengan 5 miliar
  - 35% untuk PKP di atas 5 miliar

- Your task is calculate the income tax given
  - Monthly income
  - Marital status
  - The number of dependents (jumlah tanggungan)

**Input - 1**

````python
# Write your code here

#Input Data Wajib Pajak

##monthly_income = 4750000
#marital_status = 'single'
#n_dependents = 0

monthly_income = 15_000_000
marital_status = "married"
n_dependents = 3

#Perhitungan Penghasilan Setahun
annual_income = monthly_income * 12

#Tampilkan Pendapatan Netto Tahunan
print(f"Penghasilan netto setahun : Rp {annual_income:,}")


#Variabel PTKP
PTKP_single = 54_000_000
PTKP_married = 54_000_000 + 4_500_000
PTKP_dep = PTKP_married + n_dependents*4_500_000
PTKP_maxdep = PTKP_married + 3*4_500_000

#Perhitungan dan Tampilkan PTKP
if marital_status == "married":
  if n_dependents == 0:
    PTKP = PTKP_married
    print(f"PTKP K/0 = Rp {-PTKP:,}")
  elif n_dependents >= 3:
    PTKP = PTKP_maxdep
    print(f"PTKP K/>=3 = Rp {-PTKP:,}")
  else :
    PTKP = PTKP_dep
    print(f"PTKP K/2 = Rp {-PTKP:,}")

elif marital_status == "single":
  PTKP = PTKP_single
  print(f"PTKP TK/0 = Rp {-PTKP:,}")

else :
  print("Untuk melakukan kalkulasi pajak, isi data status pernikahan dengan benar!")

#Perhitungan PKP
PKP = annual_income - PTKP

#Tampilkan PKP
print(f"Penghasilan Kena Pajak = Rp{PKP:,}")

#Variabel PPh
TPP60 = 5/100 * PKP
TPP250 = 15/100 * PKP
TPP500 = 25/100 * PKP
TPP5M = 30/100 * PKP
TPPMAX = 35/100 * PKP

#Perhitungan dan Tampilkan PPh 21
if PKP <= 60_000_000:
  print(f"PPh 21 terutang setahun = Rp {TPP60:,}")
elif PKP <=250_000_000:
  print(f"PPh 21 terutang setahun = Rp {TPP250:,}")
elif PKP <=500_000_000:
  print(f"PPh 21 terutang setahun = Rp {TPP500:,}")
elif PKP <=5_000_000_000:
  print(f"PPh 21 terutang setahun = Rp {TPP5M:,}")
else :
  print(f"PPh 21 terutang setahun = Rp {TPPMAX:,}")
````
