```python
# 3. Ambil data masing-masing grup
control = group_stats[group_stats['group'] == 'control']
treatment = group_stats[group_stats['group'] == 'treatment']

n_c = control['total_users'].values[0]
x_c = control['total_converted'].values[0]
p_c = control['conversion_rate'].values[0]

n_t = treatment['total_users'].values[0]
x_t = treatment['total_converted'].values[0]
p_t = treatment['conversion_rate'].values[0]
```

```python
# 4. Hitung pooled proportion dan standard error
p_pool = (x_c + x_t) / (n_c + n_t)
se = np.sqrt(p_pool * (1 - p_pool) * (1/n_c + 1/n_t))
```

```python
# 5. Hitung Z-score dan p-value
z_score = (p_t - p_c) / se
p_value = 1 - norm.cdf(z_score)  # uji satu sisi (one-tailed)
```

```python
# 6. Print hasil uji
print(f"\nZ-score: {z_score:.3f}")
print(f"P-value: {p_value:.3f}")

alpha = 0.05
if p_value < alpha:
    print("Kesimpulan: Tolak H0 → Penambahan deskripsi meningkatkan conversion rate secara signifikan.")
else:
    print("Kesimpulan: Gagal tolak H0 → Tidak ada bukti cukup bahwa deskripsi meningkatkan conversion rate secara signifikan.")
```
